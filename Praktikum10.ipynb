Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/RachmasariAR/Weather_Classifier/refs/heads/main/seattle_weather.csv")
df
```

```python
df = df.drop('date', axis=1)
df
```

```python
df.info()
```

Cleaning Dataset

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.describe()
```

Boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

#ukuran figure
plt.figure(figsize=(12,8))

#membuat subplot untuk setiap boxplot
plt.subplot(3,2,1)
sns.boxplot(x=df['precipitation'])
plt.title("precipitation")

plt.subplot(3,2,2)
sns.boxplot(x=df['temp_max'])
plt.title("temp_max")

plt.subplot(3,2,3)
sns.boxplot(x=df['temp_min'])
plt.title("temp_min")

plt.subplot(3,2,4)
sns.boxplot(x=df['wind'])
plt.title("wind ")

#Untuk layouting tampilan boxplot
plt.tight_layout()

plt.show()
```

Label Encoding

```python
df['weather'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['weather'] = le.fit_transform(df['weather'])
df
```

```python
df['weather'].value_counts()
```

Data Visualization

```python
weather_features = df.select_dtypes(include='float').columns.to_list()

for col in weather_features:
  sns.catplot(x="weather", y=col, kind="bar", dodge=False, height = 4, aspect = 3,  data=df, palette="Set3")
  plt.title("Rata-rata tiap column Relatif terhadap - {}".format(col))
```

```python
# Scatter plot antara dua fitur numerik
plt.figure(figsize=(8,6))
sns.scatterplot(x='temp_max', y='temp_min', hue='weather', data=df, palette='Set2')
plt.title('Scatter Plot temp_max vs temp_min')
plt.xlabel('temp_max')
plt.ylabel('temp_min')
plt.legend(title='weather')
plt.show()
```

```python
# Histogram untuk setiap fitur numerik
df[weather_features].hist(bins=30, figsize=(12,8), layout=(2,2), color='skyblue', edgecolor='black')
plt.suptitle('Histogram Setiap Fitur Numerik')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Pairplot untuk melihat hubungan antar fitur numerik
sns.pairplot(df[weather_features + ['weather']], hue='weather', palette='Set1')
plt.suptitle('Pairplot Fitur Numerik', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Melihat Korelasi antar Feature")
plt.show()
```
