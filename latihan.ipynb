```python
import pandas as pd

def show_middle(df, n):
    total_rows = len(df)
    if n >= total_rows:
        return df
    
    start = (total_rows - n) // 2
    end = start + n
    return df.iloc[start:end]
```

```python
show_middle(titanic_df, 7)
```
